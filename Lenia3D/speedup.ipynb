```python
from matplotlib import pyplot as plt
```

```python
data = open('elapse.txt').readlines()[:20]
sp = [entry.strip().split() for entry in data]
x = [int(entry[1]) for entry in sp[:10]]
cpu = [float(entry[-2]) for entry in sp[:10]]
gpu = [float(entry[-2]) for entry in sp[10:]]
```

```python
plt.title('Time(R)')
plt.xlabel('Kernel radius')
plt.ylabel('Time, s')
plt.plot(x, gpu, label='GPU')
plt.plot(x, cpu, label='CPU')
plt.legend()
plt.show()
```

```python
speedup = [cpu[i] / gpu[i] for i in range(len(gpu))]
```

```python
plt.title('Speedup(R)')
plt.xlabel('Kernel radius')
plt.ylabel('Speedup')
plt.plot(x, speedup)
plt.show()
```
