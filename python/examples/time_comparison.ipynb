```python
from matplotlib import pyplot as plt
```

```python
data = open('elapse.txt').readlines()[:30]
sp = [entry.strip().split() for entry in data]
x = [int(entry[1]) for entry in sp[:10]]
cpu = [float(entry[-2]) for entry in sp[:10]]
cuda = [float(entry[-2]) for entry in sp[10:20]]
cupy = [float(entry[-2]) for entry in sp[20:]]

plt.title('Time(R)')
plt.xlabel('Kernel radius')
plt.ylabel('Time, s')
plt.plot(x, cpu, label='CPU')
# plt.plot(x, cuda, label='CUDA')
plt.plot(x, cupy, label='CuPy')
plt.legend()
plt.show()
```

```python
speedup = [cpu[i] / cupy[i] for i in range(len(cupy))]
plt.title('Speedup(R)')
plt.xlabel('Kernel radius')
plt.ylabel('Speedup')
plt.plot(x, speedup)
plt.show()
```
